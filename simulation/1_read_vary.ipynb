```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
vin = pandas.read_csv('./data/Vcmp_sin_input_spike_0.8_2V_VDD1.csv')
vin
```

```python
spk = pandas.read_csv('./data/Vcmp_sin_output_spike_0.8_2V_VDD1.csv')
spk
```

```python
t   = torch.tensor(vin.iloc[:3000:10,0].to_numpy()).repeat(1200,1).T
VIN = torch.tensor(vin.iloc[:3000:10,1:-1:2].to_numpy().astype(np.float32))
VOUT = torch.tensor(spk.iloc[:3000:10,1:-1:2].to_numpy().astype(np.float32))

data = torch.stack([t, VIN, VOUT])
data = data[:,:,::5]
data.shape
```

```python
plt.figure(figsize=[20,5])
for k in np.linspace(0,1199,15).astype(int):
    i = k
    plt.plot(data[0,:,i], data[1,:,i], '--')
plt.legend()
plt.xlabel('t')
plt.ylabel('V_in')
```

```python
plt.figure(figsize=[20,5])
for k in np.linspace(0,1199,15).astype(int):
    i = k
    plt.plot(data[0,:,i], data[2,:,i], '--', label=f'V_in = {data[1,0,i]:.4f} V')
plt.legend()
plt.xlabel('t')
plt.ylabel('V_out')
```

```python
np.savetxt('./data/vin_vary.csv', data[1,:,:], fmt='%.8f', delimiter=',')
np.savetxt('./data/vout_vary.csv', data[2,:,:], fmt='%.8f', delimiter=',')
```
