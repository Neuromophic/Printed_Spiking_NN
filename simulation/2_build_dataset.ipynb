```python
import pandas
import torch
```

```python
vin_const = pandas.read_csv('./data/vin_const.csv', header=None)
vin_const
```

```python
vout_const = pandas.read_csv('./data/vout_const.csv', header=None)
vout_const
```

```python
vin_vary = pandas.read_csv('./data/vin_vary.csv', header=None)
vin_vary
```

```python
vout_vary = pandas.read_csv('./data/vout_vary.csv', header=None)
vout_vary
```

```python
VIN_const = torch.tensor(vin_const.to_numpy())
VIN_vary = torch.tensor(vin_vary.to_numpy())
VOUT_const = torch.tensor(vout_const.to_numpy())
VOUT_vary = torch.tensor(vout_vary.to_numpy())

VIN = torch.cat((VIN_const, VIN_vary), 1)
VOUT = torch.cat((VOUT_const, VOUT_vary), 1)

VIN.shape, VOUT.shape
```

```python
# WindowSize = 7
# X = []
# Y = []
# for sample in range(VIN.shape[1]):
#     VIN_temp = VIN[:,sample]
#     VOUT_temp = VOUT[:,sample]
#     for t in range(int(VIN.shape[0])):
#         if t < WindowSize+1:
#             continue
#         x1_temp = VOUT_temp[t-WindowSize-1:t-1]
#         x2_temp = VIN_temp[t-1].unsqueeze(0)
#         x_temp = torch.cat((x1_temp,x2_temp),0)
#         x_temp = x_temp.unsqueeze(0)
#         y_temp = VOUT_temp[t]
#         X.append(x_temp)
#         Y.append(y_temp)


X = torch.tensor(VIN.clone().detach()).float().T.unsqueeze(2)
Y = torch.tensor(VOUT.clone().detach()).float().T

# X_min = X.min(0)[0]
# X_max = X.max(0)[0]
# Y_min = Y.min(0)[0]
# Y_max = Y.max(0)[0]
# X_normalize = (X - X_min) / (X_max - X_min)
# Y_normalize = (Y - Y_min) / (Y_max - Y_min)

rand_idx = torch.randperm(X.shape[0])
train_idx = rand_idx[:int(X.shape[0]*0.8)]
valid_idx = rand_idx[int(X.shape[0]*0.8):int(X.shape[0]*0.9)]
test_idx  = rand_idx[int(X.shape[0]*0.9):]
X_train = X[train_idx]
Y_train = Y[train_idx]
X_valid = X[valid_idx]
Y_valid = Y[valid_idx]
X_test  = X[test_idx]
Y_test  = Y[test_idx]

data = {'X_train':X_train, 'Y_train':Y_train,
        'X_valid':X_valid, 'Y_valid':Y_valid,
        'X_test':X_test, 'Y_test':Y_test}

torch.save(data, f'./data/dataset.ds')

print(X.shape, Y.shape)
```
