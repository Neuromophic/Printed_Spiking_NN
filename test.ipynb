```python
import torch
import psnn
import matplotlib.pyplot as plt
```

```python
X = torch.linspace(0.1,0.1,300)
```

```python
beta = torch.tensor(0.95)
threshold = torch.tensor(1.0)

sn = psnn.Leaky(beta, threshold, init_hidden=True)
```

```python
spks = []
mems = []

# mem = sn.init_leaky()

for t in range(300):
    spk, mem = sn(X[t])
    spks.append(spk)
    mems.append(mem)
```

```python
for n, p in sn.named_parameters():
    print(n,p)
    print(p.grad)
```

```python
torch.stack(spks).squeeze().sum().backward()
for n, p in sn.named_parameters():
    print(n,p)
    print(p.grad)
```

```python
plt.figure(figsize=(30,5))
plt.plot(X)
plt.plot(spks)
plt.plot(mems)
```
